# Exp No:1 Read and Write Image and Video files

## Image File-Read

```python
#import necessary files
import cv2
import matplotlib.pyplot as plt
```

```python
# Read an image
image_path = "213253.jpg"
img = cv2.imread(image_path)
```

```python
# Display the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))    # converts the image into bgr colors and generates it
plt.title('Original Image')
plt.axis('off')
plt.show()
```

## Image File-Write

```python
# Write the image to a new file
output_image_path = "C:\\Users\\TARUN\\Downloads\\cube.jpg"
cv2.imwrite(output_image_path, img)
```

## Video file-Read

```python
# Read a video file
video_path = "C:\\Users\\TARUN\\Downloads\\sunset.mp4"
vid = cv2.VideoCapture(video_path)
```

```python
# Get the total number of frames in the video
total_frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
print(f'Total number of frames: {total_frames}')
```

```python
# Get the frames per second (fps) of the video
fps = vid.get(cv2.CAP_PROP_FPS)
fps
```

```python
# Get the width and height of the video frames
width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
print("width:",width," height:",height)
```

## Video File-Write

```python
# Create a VideoWriter object to write the video to a new file
output_video_path = 'C:\\Users\\TARUN\\Downloads\\sunset1.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can use 'XVID' or other codecs as well
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))
```
